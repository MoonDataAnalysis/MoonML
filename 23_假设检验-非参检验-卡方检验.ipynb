# 几种重要的非参检验-卡方检验

实际问题中，可能对总体分布知之甚少，难以给出总体参数的假定。这就需要用到 非参检验



## 1.卡方检验-拟合优度检验

```r
#例子： 5种啤酒，其爱好者的频数如下。求5中啤酒的爱好有无明显差异?
```

```r
X<-c(210, 312, 170, 85, 223)
chisq.test(X)
```

```r
#例子：杂交大麦麦芒类型比例  无芒：长芒：短芒 = 9:3:4， 实际观测值为335:125:160，
# 观测值是否符合理论假设？
```

```r
# H0： p1=9/16   p2=3/16   p3=4/16
chisq.test(c(335,125,160), p=c(9/16, 3/16, 4/16))
```

## 2.Pearson 卡方检验-列联表独立性检验

```r
#举例：吸烟是否与肺癌有关？ 对63位肺癌患者，以及43位正常患者调查其中的吸烟人生
```

```r
x <- array(c(60, 3, 32, 11),dim = c(2,2))  
chisq.test(x)
```

```r
#举例：工作满意程度与收入是否相关
```

```r
x<- array(c(20, 24, 80, 82, 22, 38, 104, 125, 13, 28, 81, 113, 7, 18, 54, 92),
     dim=c(4,4))
x <- t(x)
x
```
