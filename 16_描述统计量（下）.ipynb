### 分散程度的度量

```r
#常用的数据分散程度的量数有：方差、标准差、极差、四分位差、变异系数等
```

方差、标准差和变异系数

方差是描述变异性的量数，是样本相对于均值的偏差平方和的平均


样本方差开方即标准差


变异系数是数据相对分散性的一种度量，记为CV


```r
#w是一组人的体重
w <- c(75.0, 64.0, 47.4, 66.9, 62.2, 62.2, 58.7, 63.5,
       66.6, 64.0, 57.0, 69.0, 56.9, 50.0, 72.0)
w
```

```r
#计算方差
var(w)
```

```r
#计算标注差
sd(w)
```

```r
#计算变异系数
cv <- 100*sd(w)/mean(w)
cv
```

### 极差与标准误

```r
#极差
max(w) - min(w)
```

```r
sd(w)/sqrt(length(w))
```

### 分布形状的度量

偏度系数是刻画数据对称性的指标，均值对称的数据偏度系数为0，右偏态为正，左偏态为负



```r
g1<-function(x){
    m <- mean(x)
    s <- sd(x)
    n <- length(x)
    n/((n-1)*(n-2))*sum((x-m)^3)/s^2
}
```

```r
g1(w)
```

```r
plot(density(w))
```

峰度系数：数据正态分布，峰度系数为0； 分布较正态更分数，凤都系数为正，否则为负


```r
g2<-function(x){
    m <- mean(x)
    s <- sd(x)
    n <- length(x)
     ((n*(n+1))/((n-1)*(n-2)*(n-3))*sum((x-m)^4)/s^4- (3*(n-1)^2)/((n-2)*(n-3)))
}
```

```r
g2(w)
```
