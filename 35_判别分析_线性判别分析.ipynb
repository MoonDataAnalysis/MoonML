# 想象场景

1. 在市场营销中，我们根据根据一些特征数据来确定区分不同类型的客户，哪些付费哪些无感？
2. 银行想根据客户特征确定哪些可能破产，哪些比较安全？

### 欲回答这些问题——可使用判别分析

判别分析其核心思想是在P维特征空间中确定一个超平面，

能够尽可能地把不同组分开。



Fisher法思路：




```r
# 举例，我们有三种红酒的有13种化学成分指标，我们想确定一个判别规则，用13中指标判断品种
library(car)
# install.packages('rattle')
data(wine, package='rattle')
attach(wine)
head(wine)
```

```r
library(MASS)
wine.lda <- lda(Type ~ ., data=wine )
```

```r
wine.lda
```

```r
wine.lda.values <- predict(wine.lda)
wine.lda.values
```

```r
newG <- wine.lda.values$class
tab <- table(wine$Type, newG); print(tab)
cat('正确判别率=', sum(diag(prop.table(tab))), '\n')
```

```r
plot(wine.lda.values$x[,1],wine.lda.values$x[,2]) # make a scatterplot
text(wine.lda.values$x[,1],wine.lda.values$x[,2],Type,cex=0.7,pos=4,col="red") # add labels
```

### 注意：线性判别依赖一个假设：不同组的covariance matrix相同

不同时怎么办？选用二次判别

```r
wine.qda <- qda(Type ~ ., data=wine)
```

```r
wine.qda
```

```r
wine.qda.values <- predict(wine.qda)
newG <- wine.qda.values$class
tab <- table(wine$Type, newG); print(tab)
cat('正确判别率=', sum(diag(prop.table(tab))), '\n')
```
