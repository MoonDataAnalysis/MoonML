# 3.符号检验-以中位数确定正负号，确定相对位置差异

```r
# 1个样本的情况
#例：联合国统计全世界66个大城市生活花费指数，自小到大的次序排列如下：
```

```r
a=c(66, 75, 78, 80, 81, 81, 82, 83, 83, 83, 83,
  84, 85, 85, 86, 86, 86, 86, 87, 87, 88, 88,
  88, 88, 88, 89, 89, 89, 89, 90, 90, 91, 91,
  91, 91, 92, 93, 93, 96, 96, 96, 97, 99, 100,
  101, 102, 103, 103, 104, 104, 104, 105, 106, 109, 109,
  110, 110, 110, 111, 113, 115, 116, 117, 118, 155, 192)
```

```r
# 现在知道北京的指数为 99， 问：北京在中位数之上，还是在中位数之下？
#  H0：M >= 99 ; H1: M < 99  (M是样本中位数)
```

```r
# 所有样本值 - 99，得到哥数据点正负号，
# 如果 M=99，那么相减之后上下各占一半，
# 由二项分布计算出现符号，正号的个数的概率。给定显著性水平。判断是否接受H0
```

```r
binom.test(x=sum(X>99), n=length(X), p=0.5 ,alternative ="less")
```

```r
# 两配对样本的情况
# 举例：两种不同的猪饲料，猪的增重情况如下
```

```r
X <- c( 25, 30, 28, 23, 27, 35, 30, 28, 32, 29, 30, 30, 31, 16)
Y <- c( 19, 32, 21, 19, 25, 31, 31, 26, 30, 25, 28, 31, 25, 25)
```

```r
# 请回答量饲料养猪是否有显著差异？
```

```r
# 用t检验？样本太少
# 用符号检验！
# 如果两总体无显著差异，那么Xi-Yi>0, <0应该各占一半。
# 使用二项分布，确定显著性
```

```r
binom.test(sum(X<Y), length(X), p=0.5)
```
