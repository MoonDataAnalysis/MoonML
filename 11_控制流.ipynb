### switch 语句

```r
#switch语句是多分支语句
#使用格式为： switch(statement, list)
#statemnt是表达式，list是列表， 如果表达式返回值在1到length（list）之间，
#则返回列表位置的值，否则返回“NULL”
```

```r
x <- 1
switch(x, 2+2, mean(1:10), rnorm(4))
```

```r
switch(3, 2+2, mean(1:10), rnorm(4))
```

```r
switch(2, 2+2, mean(1:10), rnorm(4))
```

```r
print(switch(4, 2+2, mean(1:10), rnorm(4)))
```

### 中止与空语句

```r
#break语句用来终止循环，使程序跳出循环
#next空语句用来跳过当前内容，继续执行
#break和next会结合下面循环语句来讲
```

1.for循环
基本格式 for( i in expr_1) expr_2
expr_1 是向量表达， exper_2是执行运算

构造一个5阶希尔伯特矩阵


```r
x <- array(0,dim=c(5,5))
x
```

```r
for(i in 1:5){
    for(j in 1:5){     
        x[i,j] <- 1/(i+j-1)
    }  
}
```

```r
x
```

2.while循环
基本格式 while(cond)expr
cond成立，则执行expr

构造一个1000以内菲薄齐纳数列


```r
f <- 1; f[2] <-1; i<-1

while(f[i]+f[i+1]<1000){
    
    f[i+2] <-  f[i]+f[i+1]
    i <- i+1
}
```

```r
f
```

3.repeat 循环
基本格式 repeat expr

repeat循环依赖break跳出，例如用repeat循环算1000以内的斐波那契数列

```r
f <- 1; f[2] <-1; i<-1

repeat{
    f[i+2] <- f[i] + f[i+1]
    i <- i+1
    if(f[i] + f[i+1]>=1000) break
}
```

```r
f
```
