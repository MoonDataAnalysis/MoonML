# 直方图、经验分布图与QQ图

```r
#直方图
#常用直方图描述分布，直方图把数据取值分为若干区间（区间等间隔），
#每个区间长度为组距。考察每个数据落入区间的频率。
```

hist(x, breaks = "Sturges", freq = NULL, probability = !freq,
include.lowest = TRUE, right = TRUE,
density = NULL, angle = 45, col = NULL, border = NULL,
main = paste("Histogram of" , xname),
xlim = range(breaks), ylim = NULL,
xlab = xname, ylab,
axes = TRUE, plot = TRUE, labels = FALSE,
nclass = NULL, ...)

其中x是样本向量， breaks是直方图组距

```r
#核密度估计（kernal density estimate）函数
#用于已知样本，估计密度
```

density(x, bw = "nrd0", adjust = 1,
kernel = c("gaussian", "epanechnikov", "rectangular",
"triangular", "biweight", "cosine", "optcosine"),
window = kernel, width,
give.Rkern = FALSE,
n = 512, from, to, cut = 3, na.rm = FALSE)

```r
#例：画出15位学生体重的直方图
w <- c(75.0, 64.0, 47.4, 66.9, 62.2, 62.2, 58.7, 63.5,66.6, 64.0,
       57.0, 69.0, 56.9, 50.0, 72.0)
w
```

```r
hist(w, freq = F)
```

```r
hist(w, freq = F)
lines(density(w), col = "blue")
```

```r
#我添加一个正态分布的概率密度曲线
hist(w, freq = F)
lines(density(w), col = "blue")
 
lines(x=44:76, y=dnorm(44:76, mean(w), sd(w)), col = "red")
```

```r
#经验分布
#直方图适合总体为连续分布的情况，若要估计总体分布函数F（x）,
#可用经验分布法（empirical distribution function）
```

什么叫经验分布函数


使用ecdf(x)函数，其中x是观察得到的数值向量，

```r
#例：汇出15位学生体重的经验分布图和相应的正态分布图
w
```

```r
ecdf(w)
```

```r
plot(ecdf(w),verticals = T, do.points = T)
```

```r
plot(ecdf(w),verticals = T, do.points = F)
lines(44:78, pnorm(44:78, mean(w), sd(w)))
```

```r
# QQ图
# QQ图可以帮助我们鉴别样本是否近似于某种类型的分布
#使用qqnorm函数和qqline函数绘制QQ图和相应直线
```

样本数据如果近似正态分布的话，数据点会贴近下面这条直线

其中theta是标准差，miu是截距

```r
#例：画出15位学生体重的QQ图，判断是否为正态
w
```

```r
qqnorm(w)
qqline(w)
```
