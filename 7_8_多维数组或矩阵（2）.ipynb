### 数组的四则运算

```r
#可以对数组之间进行四则运算，运算的数组一般应是的dim属性
```

```r
A <- matrix(1:6, nrow=2, byrow=T)
print(A)
```

```r
B <- matrix(1:6, nrow=2)
print(B)
```

```r
C <- matrix(c(1,2,2,3,3,4),nrow=2)
print(C)
```

```r
D <- 2*C+A/B
print(D)
```

```r
#可见数值的加减乘除只是对应位置上元素的运算
```

```r
#形状不一致的数组（向量）也可以进行四则运算，把小数组（向量）的数据循环使用并匹配
x1 <- c(100,200)
x2 <- 1:6
print(x1)
print(x2)
```

```r
x1+x2
```

```r
x3 <- matrix(1:6, nrow = 3)
print(x3)
print(x1)
```

```r
x1 + x3
```

```r
#如果向量按列匹配数组，匹配不上时，R会给出警告
x2 <-1:5
print(x2)
print(x1)
```

```r
x1 + x2
```

### 矩阵的运算

```r
#1.转置运算，使用函数t()
A <-matrix(1:6, nrow=2)
print(A)
```

```r
t(A)
```

```r
#求方阵的行列式,使用函数det（）
B <- matrix(1:4, ncol=2)
print(B)
```

```r
det(B)
```

```r
#求向量的内积
#n维向量，可以看做n*1的矩阵，或1*n的矩阵。x，y是相同维数的向量，x%*%y表示x与y的内积
```

```r
x <- 1:2
print(x)
```

```r
y<-2*1:2
print(y)
```

```r
x %*% y
```

```r
#求向量的外积（叉积）
#x %o% y表示x与y的外积
x <- 1:5
y <- 2*1:5
print(x)
print(y)
```

```r
x %o% y
```

### 矩阵的乘法

```r
#如果矩阵A和B 具有相同的维数，A*B是对应位置元素的乘积，
#A%*%B表示数学意义上矩阵的乘法
```

```r
A <- array(1:9, dim= c(3,3))
print(A)
```

```r
B <- array(9:1, dim=c(3,3))
print(B)
```

```r
A*B
```

```r
A %*% B
```

```r
#生成对角阵 和 矩阵取对角运算
#使用函数diag()。  当v是一个向量，diag（v）表示以v为对角线的对角阵。
#当M是一个矩阵时，diag（M）表示的是取M对角线上的向量
```

```r
v <- c(1,4,5)
print(v)
```

```r
diag(v)
```

```r
M <- array(1:9, dim=c(3,3))
print(M)
```

```r
diag(M)
```

```r
#解线性方程组和求矩阵的逆矩阵
#若求解线性方程组Ax =b，其命令是solve(A,b)
```

```r
A<- t(array(c(1:8,10),dim=c(3,3)))
b = c(1,1,1)
print(A)
print(b)
```

```r
solve(A,b)
```

```r
#若欲求解矩阵A的逆
solve(A)
```

```r
#求矩阵的特征值和特征向量，使用eigen（）函数
```

```r
print(A)
```

```r
Sm <- crossprod(A,A) #crossprod(A,B)表示t(A) %*% B
print(Sm)
```

```r
eigen(Sm)
```

```r
#其中$values是Sm的特征值构成的向量
#$vectors是Sm的特征向量构成的矩阵
```

矩阵的奇异值分解
svd（A）即对矩阵A做奇异值分解


```r
print(A)
```

```r
svdA <- svd(A)
print(svdA)
```

```r
#$d表示矩阵A的奇异值，即矩阵D对角线上的元素。
#$u对应的是正交阵U
#$v对应的是正交阵V
```

```r
svdA$u %*% diag(svdA$d) %*% t(svdA$v)
```

```r
A <- matrix(1:6, nrow=2)
print(A)
```
