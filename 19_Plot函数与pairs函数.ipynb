# Plot函数与pairs函数

plot（）函数

plot可以绘制数据的散点图、曲线图等，plot函数有以下四种使用方法

（1） plot(x, y)
x和y是向量， 对应x轴与y轴，生成散点图

```r
#给到5个坐标点（两维）的，分别是（2,4），（3,4），（5,4），（6,5），（2,5）
x <- c(2,3,5,6,2)
y <- c(4,4,4,5,5)
plot(x,y)
```

（2） plot(x)
x是时间序列，则产出x关于下标的散点图


```r
#一个学生从开学到期末，10次成绩的
s <- c(97,98,99,95,94,96,97,99,100,98)
plot(s)
```

（3）plot(f)

   plot(f, y)
   
其中f是因子， y是数值向量；第一种格式生成f的直方图；第二种格式生成y关于f水平的箱线图

```r
#例如：有四种配方不同的材料A1， A2, A3, A4 生产出来的元件，各自使用寿命数据如下
y<-c(1600, 1610, 1650, 1680, 1700, 1700, 1780, 1500, 1640,1400, 
     1700, 1750, 1640, 1550, 1600, 1620, 1640, 1600,1740, 1800, 
     1510, 1520, 1530, 1570, 1640, 1600)
f<-factor(c(rep(1,7),rep(2,5), rep(3,8), rep(4,6)))
```

```r
plot(f,y)
plot(f)
```

(4) plot(y ~ expr, data=df)

df是数据框， y是任一对象， expr是对象名的表达式（a+b+c）

```r
df <- data.frame(
Age = c(13, 13, 14, 12, 12, 15, 11, 15, 14, 14, 14,
        15, 12, 13, 12, 16, 12, 11, 15 ),
Height=c(56.5, 65.3, 64.3, 56.3, 59.8, 66.5, 51.3,
         62.5, 62.8, 69.0, 63.5, 67.0, 57.3, 62.5,
         59.0, 72.0, 64.8, 57.5, 66.5),
Weight=c( 84.0, 98.0, 90.0, 77.0, 84.5, 112.0,
         50.5, 112.5, 102.5, 112.5, 102.5, 133.0,
         83.0, 84.0, 99.5, 150.0, 128.0, 85.0,112.0)
)
df
```

```r
attach(df)
```

```r
plot(~ Age + Height)
```

```r
plot(Weight ~ Age + Height)
```

```r
# pairs()函数显示多变量数据的配对关系
head(df)
```

```r
pairs(df)#结果与plot(x)相同
```
