# 递归函数

```r
#R语言可以在函数自身内定义函数本身
```

例：用递归函数计算数值积分，下面[1:5]的定积分，精度要求10-6




```r

area <-function(f,a,b,eps=1e-6, lim=10){
    "
参数;
f是被积函数
a,b是基本的两端点
eps是精确度

返回：
area函数是主程序，首先用梯形公式计算积分近似值，然后调用fun1
fun是子程序，其意义是：将区间对分，采用复化求积公式，
如果计算值和上次计算值相差精度小于eps，或lim=0，停止计算，否则分别调用自身函数
    "
    fun1 <- function(f,a,b,fa,fb,a0,eps,lim,fun){
        d<- (a+b)/2;
        h<- (b-a)/4;
        fd <- f(d)
        a1 <- h*(fa + fd);
        a2 <- h*(fd + fb)
        if(abs(a0 -a1 -a2)<eps || lim ==0){return(a1+a2)}
        else{return(fun(f,a,d,fa,fd,a1,eps,lim-1,fun)
                   +fun(f,d,b,fd,fb,a2,eps,lim-1,fun))
            }          
    }
    fa <- f(a); fb<-f(b); a0 <-((fa+fb)*(b-a))/2
    fun1(f,a,b,fa,fb,a0,eps,lim,fun1)
}
```

```r
f <- function(x){1/x}
```

```r
area(f,1,5)
```
