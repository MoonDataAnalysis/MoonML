# 秩相关检验

之前讲到的Pearson相关检验，要求样本来自正态分布总体。那如果不能确定，或者根本不是正态分布呢？
## 1.Spearman秩相关检验

```r
#使用cor.test(X,Y, method='spearman')

#举例，有六人参加表演竞赛，两人评定是否一致
#两评定人打分结果如下
x<-c(1,2,3,4,5,6); y<-c(6,5,4,3,2,1)
```

```r
cor.test(x,y,method = 'spearman')
```

## 2.kendall相关检验


```r
#使用cor.test(X,Y, method='kendall')
x
y
```

```r
cor.test(x,y,method = 'kendall')
```

# 符号秩次检验法——Wilcoxon检验

```r
#使用wilcox.test（）或binom.test()

#例：某电池厂电池中位数为140安倍，取新生产的一批电池20个，寿命如下：
a  <- c(137.0, 140.0, 138.3, 139.0 ,144.3, 139.1 ,141.7 ,137.3 ,133.5 ,138.2,
  141.1, 139.2, 136.5, 136.5, 135.6, 138.0, 140.9, 140.6, 136.3, 134.1)
#该样本寿命是否显著偏小？
```

```r
wilcox.test(a, mu=140, alternative="less",
            exact=FALSE, correct=FALSE, conf.int=TRUE)
```

```r
#还可以用于检验量样本差异

#两块天，一部分用复合肥，一部分用化肥，两者产量如下
x<-c(459, 367, 303, 392, 310, 342, 421, 446, 430, 412)
y<-c(414, 306, 321, 443, 281, 301, 353, 391, 405, 390)
#复合肥是否好于化肥？
```

```r
wilcox.test(x, y, alternative = "greater", paired = TRUE)
```
