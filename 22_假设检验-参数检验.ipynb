## 假设检验的整体流程

## 假设检验两类错误

## 1.均值差异的检验-单样本情况

```r
#例：16只元器件，寿命如下
a <- c(159, 280, 101, 212, 224, 379, 179, 264,
  222, 362, 168, 250, 149, 260, 485, 170)
```

```r
#该厂平均寿命为225，该样本与厂平均寿命是否有差别？
#  u0=225,   H0: u = u0,  H1: u != u0(双尾)
```

```r
#（1）知道总体标准差为5
library(BSDA)
z.test(a, mu=225, sigma.x = 50, alternative = 'two.sided')
```

```r
#(2)不知道总体标准差
t.test(a,mu=225, alternative = 'two.sided')
```

## 2.均值差异的检验-双样本差异比较

```r
#现在有另外一批元件，寿命如下
b <- c(225, 234, 236, 254, 247, 228, 230, 214, 267, 278)
```

```r
#两批元件是否有显著差异？
#   H0： 两样本无差，u1=u2,    H1:有差， u1!=u2
```

```r
#(1) 总体标准差已知，两总体都是50
z.test(a, b, sigma.x = 50, sigma.y = 50, alternative = 'two.sided')
```

```r
#(2) 总体标准差未知,但知道来自一个总体
t.test(a,b, var.equal = T,alternative = 'two.sided')
```

## 3.均值差异的检验-双样本成对比较

```r
#1批元件，使用前质量和使用后质量是否有差
X<-c(78.1,72.4,76.2,74.3,77.4,78.4,76.0,75.5,76.7,77.3) #使用前
Y<-c(79.1,81.0,77.3,79.1,80.0,79.1,79.1,77.3,80.2,82.1)  #使用后
```

```r
#(1) 不知道总体标注差
# H0：使用前后无差  H1:使用前后有差
```

```r
t.test(Y-X, alternative = 'two.sided')
```

```r
#（2）知道样本来自标准差为50的总体
```

```r
z.test(Y-X, sigma.x = 5, sigma.y = 5, alternative = 'two.sided')
```

## 4.二项分布总体的假设检验

```r
#数据不是连续的，而是二分类的
```

例如：有一批种子，发芽率 p0=0.85，现随机抽取500粒。用药剂浸泡，结果有445粒发芽，请问药剂对发芽率是否有效？ 

```r
#假设 H0: 无效，p<=p0=0.85， H1：有效，p>p0
binom.test(x=445, n=500, p=0.85, alternative = 'greater')
```
