```r
test<-scale(test)
test
```

```r
ca<-cancor(test[,1:3],test[,4:6])
```

```r
ca
```

```r
U<-as.matrix(test[, 1:3])%*% ca$xcoef
V<-as.matrix(test[, 4:6])%*% ca$ycoef
U
V
```
